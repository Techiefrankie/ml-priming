## Visualising Data With Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [i * np.random.randint(1, 11) for i in np.arange(1, 11)]
data
```

```python
explosions = np.zeros(len(data))
min_pos = np.argmin(data)
max_pos = np.argmax(data)
explosions[min_pos] = 0.4
explosions[max_pos] = 0.8
plt.pie(data, labels=[i for i in data], autopct='%.2f%%', explode=explosions)
```

### Scatter Plot

```python
x = np.linspace(0, 10, 50)
y = x ** 3

plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.datasets import make_blobs
data = make_blobs()
points = data[0]
x,y = points[:, 0], points[:, 1]
plt.plot(x,y)
```

```python
plt.scatter(x,y, c=data[1])
```

```python
x = np.linspace(-np.pi * 2, np.pi * 2, 721)
y = np.sin(x)
plt.plot(x, y, c='Green')
```

### Line Styles

```python
styles = ['solid', 'dotted', 'dashed', 'dashdot', 'None']
for index, style in enumerate(styles):
    plt.plot(np.arange(10) + index, linestyle=style, linewidth=5, dash_capstyle='round')
```

```python
x = np.random.randint(1, 10, 10)
plt.plot(x, solid_joinstyle='bevel', lw=10)
```

### Markers

```python
for index, symbol in enumerate(list('osv<>+^8*pP')):
    plt.scatter([0 + index], [0], marker=symbol)
```

```python
plt.scatter([0], [0], marker=(5,0,0))  # markers are plottes as polygons
```

```python
rand = np.random.randint(1,10,10)
plt.plot(rand, marker=(3,0,0))
plt.plot(rand, marker=(3,0,0), ls='None')
```

### Legend

```python
data1 = [np.sin(x) for x in np.arange(1,11)]
data2 = [np.cos(x) for x in np.arange(1,11)]
data3 = [np.tan(x) for x in np.arange(1,11)]
plots = []
plots.append(data1)
plots.append(data2)
plots.append(data3)

handles = []
for plot in plots:
    line = plt.plot(plot, label='x')
    handles.append(line)
handles[1].set_lw(5)
handles[1].set_ls('dotted')
plt.legend(loc=8, ncol=2, facecolor='#ffcccc', fontsize='large', shadow=True, title='Trignometry')
```
